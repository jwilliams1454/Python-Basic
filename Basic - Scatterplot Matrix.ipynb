```python
# Python scatter matrix with Pandas
# https://www.marsja.se/pandas-scatter-matrix-pair-plot/
import numpy as np
import pandas as pd

#use NumPy to create 3 variables (x1, x2, and x3)
np.random.seed(134)                     
N = 1000                              
 
x1 = np.random.normal(0, 1, N)                        
x2 = x1 + np.random.normal(0, 3, N)              
x3 = 2 * x1 - x2 + np.random.normal(0, 2, N)
```

```python
# Example 1

#create a Pandas dataframe from a dictionary
df = pd.DataFrame({'x1':x1,
                   'x2':x2,
                   'x3':x3})

df.head()
```

```python
# Creating the scatter matrix:
# The diagonal shows the distribution of the three numeric variables of our example data.
# In the other cells of the plot matrix, we have the scatterplots (i.e. correlation plot) of each variable combination of our dataframe
pd.plotting.scatter_matrix(df)
```

```python
# Example 2

# Changing the number of bins of the scatter matrix in Python:
pd.plotting.scatter_matrix(df, hist_kwds={'bins':30})
```

```python
# Example 3

# Scatter matrix with Pandas and density plots:
pd.plotting.scatter_matrix(df, diagonal='kde')
```
