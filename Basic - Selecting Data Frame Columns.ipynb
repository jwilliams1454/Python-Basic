```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
import re
X = load_wine()
df = pd.DataFrame(X.data, columns = X.feature_names)
df.head()
```

```python
# Selecting columns based on their name
# Put the string name of the column in brackets
# Returns a pandas series.
df['hue']
```

```python
# Passing a list in the brackets lets you select multiple columns 
df[['alcohol','hue']]
```

```python
# Selecting a subset of columns found in a list
# Not sure what this is doing
df[df.columns[df.columns.isin(['alcohol','hue','NON-EXISTANT COLUMN'])]]
```

```python
# Everything EXCEPT these columns
df[df.columns.difference(['alcohol','hue'])]
```

```python
# Selecting columns based on their data type
df.loc[:,(df.dtypes=='float64').values]
```

```python
# Select columns with names containing "al" (e.g., ALcohol, mALic_acid, etc.)
df.loc[:,['al' in i for i in df.columns]]
```

```python
# Using a wild card string
df.loc[:,[True if re.search('flava+',column) else False for column in df.columns]]
```

```python
# which is a complicated way to do...
df.loc[:,['flava' in i for i in df.columns]]
```

```python
# Column name starts with AL
df.loc[:,df.columns.str.startswith('al')]
```

```python
# Column name ends with OIDS
df.loc[:,df.columns.str.endswith('oids')]
```

```python
# Selecting columns if any row of a column meets a condition
# In this case any values greater than 14
df.loc[:,[(df[col] > 14).all() for col in df.columns]]
```

```python
# If all the values in the column meet a condition
# In this case the mean is greater than 7
df.loc[:,[(df[col].mean() > 7) for col in df.columns]]
```
